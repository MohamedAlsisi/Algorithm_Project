```python
def pipelinebinarysearch(n,k):
    def Sum(number):
      number = number -1
      return k*(k+1)//2 - number*(number+1)//2      
    if n == 1:
      return 0
    elif k >= n:
      return 1  
    elif ( (k*(k+1)//2) -1 -(k-2) )  < n:
      return -1
    k = k-1
    n = n-1
    start = 1
    end = k
    while start < end:
      middle = start + (end-start)//2
      summation = Sum(middle)
      if summation == n:
        return k - middle + 1
      elif summation > n:
        start = middle + 1
      else:
        end = middle
    return k- start + 2  #the plus 2 is put, 1 for the counting to include the start element, and 1 to include the number before the start element as the loop exits when the sum is just less than the needed number of houses so we need one more splitter 
n, k = input().split()
n = int(n)
k = int(k)
print(pipelinebinarysearch(n,k))
```

```python
############  https://codeforces.com/contest/287/submission/142065380
def pipelinebruteforce(n,k):
     if n == 1:
        return 0
  if k >= n:
      return 1
  if ((k*(k+1)//2) -1 -(k-2)) < n:
      return -1
  splitters = 0
  summation = k
  for i in range (k-1, 1, -1):
      summation = summation + i -1
      if summation >= n:
          splitters = (k-i)+1
          break
  return splitters
n, k = input().split()
n = int(n)
k = int(k)
print(pipelinebruteforce(n,k))
```

$$T(f(n, k)) = 2 + 3k$$

$$T(f(n, k)) = k$$

```python
import random
import matplotlib.pyplot as plt
splitters_random = []
houses_random = []
for i in range(1, 1000):
  splitters_sample = random.randint(2,100000)
  houses_sample = random.randint(1,100000000)
  splitters_random.append(splitters_sample)
  houses_random.append(houses_sample)
splitters_random
houses_random
```

```python
import time
times1 = []
times2 = []

for i in range(len(houses_random)):
    time1 = time.time()
    pipelinebinarysearch(houses_random[i],splitters_random[i])
    time1 = time.time() - time1
    
    time2 = time.time()
    pipelinebruteforce(houses_random[i],splitters_random[i])
    time2 = time.time() - time2
    
    
    times1.append(time1)
    times2.append(time2)
```

```python
plt.plot(times2, label='Brute Force')
plt.legend()
plt.show()
```
