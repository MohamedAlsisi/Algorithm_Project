## Problem
*  C. Number of Ways #266 Div.2 // Mohamed Alsisi 
* Proble : https://codeforces.com/problemset/problem/466/C
* codeforce submission : https://codeforces.com/problemset/submission/466/142413016


```python
"""
    
    C. Number of Ways #266 Div.2 // Mohamed Alsisi 
    Proble : https://codeforces.com/problemset/problem/466/C
    codeforce submission : https://codeforces.com/problemset/submission/466/142413016
"""
def ways1(k):
    """
    @params:
    k: list containing all elements
    
    @returns:
    ans: number of all possible combinations of three subarrays in k such that
         S1 = S2 = S3
    """
    
    total_sum = 0
    count_s = 0
    ans = 0
    sk = sum(k)
    if sk % 3 != 0:
        return 0
    S = sk//3
    for i in range(len(k)-1): 
        total_sum += k[i]
        if total_sum == 2*S:
            ans += count_s
        if total_sum == S:
            count_s += 1
    return ans
```

$f(n) → ways1(k)$


---

$$T(f(n)) = 4 + \sum_{i=0}^{n-1}n[i] + 3 + \sum_0^{n-2}5 + 1$$

$$T(f(n)) = 8 + \sum_{i=0}^{n-1}n[i] + \sum_0^{n-2}5$$

$$T(f(n)) = 8 + n + 5(n-1) → 8 + n + 5n - 5$$

$$T(f(n)) = 3 + 6n$$

$$ \text{Suppress lower order terms and constants}$$

$$T(f(n)) = n$$


```python
"""
    codeforce submission :
    https://codeforces.com/problemset/submission/466/142020961

"""

def ways2(arr):
    """
    @params:
    arr: list containing all elements
    
    @returns:
    
    ans: number of all possible combinations of three subarrays in k such that S1 = S2 = S3
    """
    
    
    total_sum =  sum(arr)
    count_s = 0
    ans = 0
    temp = 0
    s = total_sum // 3

    if total_sum % 3 != 0:
        #Can't  array into three contiguous parts with the same sum
        return 0
    else:
        fword = [0 for i in arr]
        bword = [0 for i in arr]
        # sum for forword
        for i in range(len(arr)-1):
            temp += arr[i]
            if temp == s:
                fword[i] = 1
        temp = 0
        for i in range(len(arr)-1,-1,-1):
            temp += arr[i]
            if temp == s:
                bword[i] = 1
        #print(bword,'\n')

        for i in range(len(arr)-1,0,-1):
            #print(i,'\n')
            bword[i - 1] = bword[i - 1] + bword[i]
        #print(bword,'\n')

        for i in range(len(arr)-2):
            if fword[i] == 1:
                ans += bword[i + 2]
        return ans
```

$f(n) → ways2(k)$


---

$$ T(f(n)) = \sum_{i=0}^{n-1}n[i] + 8 + \sum_{i=0}^{n-1}3 + 1 + \sum_{i=0}^{n-1}3 + \sum_{i=1}^{n-1}1 + \sum_{i=0}^{n-2}2 + 1$$

$$ T(f(n)) = n + 8 + 3n + 1 + 3n + (n-1) + 2(n-1) + 1$$

$$ T(f(n)) = n + 10 + 3n + 3n + (n-1) + 2(n-1)$$

$$ T(f(n)) = n + 10 + 3n + 3n + n - 1 + 2n - 2$$

$$ T(f(n)) = n + 7 + 3n + 3n + n + 2n$$

$$ T(f(n)) = 10n + 7 $$

$$ \text{Suppress lower order terms and constants}$$

$$ T(f(n)) = n $$

```python
"""
    codeforce submission :
   https://codeforces.com/problemset/submission/466/142019129
    O(n^2)
"""
def ways3(arr):
    
    total_sum =  sum(arr)
    count_s = 0
    ans = 0
    temp = 0
    s = total_sum // 3

    if total_sum % 3 != 0:
        #Can't  array into three contiguous parts with the same sum
        return 0
    else:
        fword = [0 for i in arr]
        bword = [0 for i in arr]
        # sum for forword
        for i in range(len(arr)-1):
            temp += arr[i]
            if temp == s:
                fword[i] = 1
        temp = 0
        for i in range(len(arr)-1,-1,-1):
            temp += arr[i]
            if temp == s:
                bword[i] = 1

        for i in range(len(arr)-1):
            for j in range(len(arr)-1, i+1 ,-1):
                if fword[i] == bword[j] and fword[i] == 1:
                    ans += 1
        return ans
```
